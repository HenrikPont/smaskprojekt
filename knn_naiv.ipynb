```python
import numpy as np
import pandas as pd

from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, f1_score
from imblearn.pipeline import Pipeline

import features
```

```python

# Select features
selected_features = [
    "hour_of_day", "day_of_week", "month", "holiday", "weekday", "summertime",
    "temp", "dew", "humidity", "precip", "snowdepth", "windspeed",
    "cloudcover", "visibility"
]

# Load train/test split
X_train, X_test, Y_train, Y_test = features.train_test_data(selected_features, random_state=0)

# Naive kNN pipeline (no scaling, default parameters)
pipe = Pipeline([
    ("knn", KNeighborsClassifier(metric="euclidean"))  # defaults: n_neighbors=5, weights='uniform', metric='minkowski'
])

# Fit model
pipe.fit(X_train, Y_train)

# Make predictions
preds = pipe.predict(X_test)

# Evaluate
print("F1 Score:", f1_score(Y_test, preds))
print("\nClassification Report:")
print(classification_report(Y_test, preds))

# Confusion matrix
print("\nConfusion Matrix:")
print(pd.crosstab(preds, Y_test, rownames=["Predicted"], colnames=["Actual"]))
```

```python

# Select features
from sklearn.model_selection import GridSearchCV


selected_features = [
    "weather_pca1", "weather_pca2", "bad_conditions", "summertime", "hour_cos", "hour_sin", "day_sin","day_cos"
]


# Load train/test split
X_train, X_test, Y_train, Y_test = features.train_test_data(selected_features, random_state=0)

# Naive kNN pipeline (scaling, default parameters)
pipe = Pipeline([
    ("scaler", StandardScaler()),
    ("knn", KNeighborsClassifier())  # defaults: n_neighbors=5, weights='uniform', metric='minkowski'
])

param_grid = {
    "knn__n_neighbors": [3, 5, 7, 9, 11, 15, 21, 23, 25, 27, 29, 31, 33, 35, 37, 39],
    "knn__weights": ["uniform", "distance"],
    "knn__metric": ["euclidean", "manhattan", "minkowski"]
}

grid = GridSearchCV(
    pipe,
    param_grid,
    cv=5,
    scoring="f1" # Use F1-score for imbalanced classification
)

grid.fit(X_train, Y_train)

print("Best parameters:", grid.best_params_)
print("Best CV F1:", grid.best_score_)

# Make predictions using the best estimator from grid search
preds = grid.predict(X_test)

# Evaluate
print("F1 Score:", f1_score(Y_test, preds))
print("\nClassification Report:")
print(classification_report(Y_test, preds))

# Confusion matrix
print("\nConfusion Matrix:")
print(pd.crosstab(preds, Y_test, rownames=["Predicted"], colnames=["Actual"]))

```

```python

# Select features
from sklearn.model_selection import GridSearchCV


selected_features = [
    "weather_pca1", "weather_pca2", "bad_conditions", "summertime", "hour_cos", "hour_sin", "day_sin","day_cos"
]


# Load train/test split
X_train, X_test, Y_train, Y_test = features.train_test_data(selected_features, random_state=0)

# Naive kNN pipeline (scaling, default parameters)
pipe = Pipeline([
    ("scaler", StandardScaler()),
    ("knn", KNeighborsClassifier())  # defaults: n_neighbors=5, weights='uniform', metric='minkowski'
])

param_grid = {
    "knn__n_neighbors": [3, 5, 7, 9, 11, 15, 21, 23, 25, 27, 29, 31, 33, 35, 37, 39],
    "knn__weights": ["uniform", "distance"],
    "knn__metric": ["euclidean", "manhattan", "minkowski"]
}

grid = GridSearchCV(
    pipe,
    param_grid,
    cv=5,
    scoring="f1" # Use F1-score for imbalanced classification
)

grid.fit(X_train, Y_train)

print("Best parameters:", grid.best_params_)
print("Best CV F1:", grid.best_score_)

best_model = grid.best_estimator_
probs = best_model.predict_proba(X_test)[:, 1]

preds = (probs >= 0.4).astype(int)
# Evaluate
print("F1 Score:", f1_score(Y_test, preds))
print("\nClassification Report:")
print(classification_report(Y_test, preds))

# Confusion matrix
print("\nConfusion Matrix:")
print(pd.crosstab(preds, Y_test, rownames=["Predicted"], colnames=["Actual"]))

```
